```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

plt.show()
```

#### Anotación básica
* El primer parámetro es el texto de la anotación
* xy representa la coordenada a la que apunta la anotación
* xytext es la posición dónde comienza el texto de la anotación. Por defecto (xy) si se indica.
* arrowprops permite configurar el objeto FancyArrowPatch object para la flecha a dibujar en la anotación. 
Más información los parámetros para configurar FancyArrowPatch:
https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.FancyArrowPatch.html

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate('min value', 
            xy=(1, 2), 
            xytext=(1.5,2.0),
            arrowprops=dict(color='g'))

plt.show()
```

#### Al cambiar la posición del texto, cambia la orientación de la flecha

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate('min value', 
            xy=(1,2), 
            xytext=(1,3),
            arrowprops=dict(color='g'))

plt.show()
```

#### Cambiando la configuración de la flecha

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate('min value', 
            xy=(1,2), 
            xytext=(1,3),
            arrowprops=dict(facecolor='y', edgecolor='green', alpha=0.3))

plt.show()
```

#### Cambiando la configuración del texto

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate('min value', 
            xy=(1,2), 
            xytext=(1,3),
            arrowprops=dict(facecolor='y', edgecolor='green', alpha=0.3),
            color='red',style='italic')

plt.show()
```

#### Añadiendo un punto a la anotación

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate('significant point', 
            xy=(2, 4), 
            xytext=(2.0, 2.5),
            arrowprops=dict(color='g'))

ax.plot([2], [4], "ro")

plt.show()
```

#### Parámetro shrink
FancyArrowPatch permite reducir el tamaño de la flecha, especificando un factor de reducción. Aquí reducimos la flecha al 50%.

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate('significant point', 
            xy=(2, 4), 
            xytext=(2.0, 2.5),
            arrowprops=dict(color='g', shrink=0.5))

plt.show()
```

### Anotaciones avanzadas

```python
import numpy as np
```

#### Definimos unos datos para las gráficas

```python
np.random.seed(0)

x1 = -1 + np.random.randn(100)
y1 = -1 + np.random.randn(100)

x2 = 1 + np.random.randn(100)
y2 = 1 + np.random.randn(100)
```

#### Mostramos esos gráficos en una figura

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

plt.show()
```

#### Definir una caja de texto
La caja es un objeto FancyBboxPatch:
https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.FancyBboxPatch.html

Definimos propiedades de la caja:
* boxstyle 'square' representa un rectángulo
* pondremos el color de la caja a blanco con transparencia de 0.5

#### Añadir el texto dentro de la caja
* Los primeros dos parámetros de ax.text() representan las coordenadas del centro ya que hemos fijado  ha and va a "center"
* El tamaño de la fuente de nuestro texto está fijado a 20
* Usamos la definición de la caja definida para indicar que será la que rodea al texto

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

bbox_props = dict(boxstyle='square', 
                  facecolor='w', 
                  alpha=0.5)

ax.text(-2, -2, 
        'Muestra A', 
        ha='center', 
        va='center', 
        size=20,
        bbox=bbox_props)

ax.text(2, 2, 
        'Muestra B', 
        ha='center', 
        va='center', 
        size=20,
        bbox=bbox_props)
```

### Texto dentro de una flecha
Podemos definir la "caja" como una flecha

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

ax.text(-2, -2, 
        'Muestra A', 
        ha='center', 
        va='center', 
        size=20,
        bbox=bbox_props)

ax.text(2, 2, 
        'Muestra B', 
        ha='center', 
        va='center', 
        size=20,
        bbox=bbox_props)

arrow_bbox_props = dict(boxstyle='rarrow', 
                        facecolor='#EBF5FB', 
                        edgecolor='b', 
                        linewidth=2,
                        alpha=0.7)
ax.text(0, 0, 
        'Dirección', 
        ha='center', 
        va='center', 
        rotation=60,
        size=15,
        bbox=arrow_bbox_props)

plt.show()
```

### Diferentes formas de cajas
Algunas formas disponibles:  larrow, rarrow, darrow (flecha bidireccional), circle, round (rectángulo redondeado), sawtooth, etc 

Más información sobra las formas disponibles:
https://matplotlib.org/stable/gallery/shapes_and_collections/fancybox_demo.html

Algunos ejemplos

```python
sawtooth_bbox_props = dict(boxstyle='sawtooth', 
                           facecolor='#ABEBC6', 
                           alpha=0.8)
```

```python
round4_bbox_props = dict(boxstyle='round', 
                         facecolor='#EBDEF0', 
                         alpha=0.8)
```

```python
arrow_bbox_props = dict(boxstyle='darrow', 
                        facecolor='#FDEDEC', 
                        alpha=0.5)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

ax.text(-2, -2, 
        'Sample A', 
        ha='center', 
        va='center', 
        size=20,
        bbox=sawtooth_bbox_props)

ax.text(2, 2, 
        'Sample B', 
        ha='center', 
        va='center', 
        size=20,
        bbox=round4_bbox_props)

ax.text(0, 0, 
        'Direction', 
        ha='center', 
        va='center', 
        rotation=60,
        size=15,
        bbox=arrow_bbox_props)
```
