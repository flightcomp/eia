Para leer la imagen, utilizamos la función <code>imread</code> (<a href="https://scikit-image.org/docs/stable/api/skimage.io.html">documentación de la función</a>) del módulo <code>io</code> de la librería <code>scikit-image</code>.

<h2> Mostrar una imagen</h2>

Para mostrar una imagen por pantalla vamos a utilizar varias funciones de <code>matplotlib.pyplot</code> (<a href="https://matplotlib.org/stable/api/pyplot_summary.html">documentación</a>):

<ul>
    <li><code>figure</code>: Con ella crearemos una figura, dentro de la cual, mostraremos las imágenes. Podemos indicar el tamaño de la figura con el parámetro <code>figsize</code> (las unidades son pulgadas).</li>
    <li><code>imgshow</code>: Muestra una imagen en la figura.</li>
    <li><code>show</code>: Muestra la figura por pantalla.</li>
    <li><code>axis</code>: si queremos ocultar los ejes, llamaremos a esta función con el parámetro 'off'.</li>
</ul>

A continuación, un par de ejemplos de uso de estas funciones con distintos parámetros:

```python
plt.figure(figsize=(10,5)) # width = 10, height = 5 
plt.imshow(image1)
plt.show()
```

Ahora sin mostrar los ejes:

```python
plt.figure(figsize=(10,5))
plt.imshow(image1)
plt.axis('off') #desactivamos los ejes
plt.show()
```

<h2> Mostrar varias imágenes en una misma figura</h2>

Podemos mostrar varias imágenes en una misma figura. Para ello, vamos a utilizar la función <code>subplot</code>(<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html">documentación de la función</a>) con tres parámetros:
<ul>
    <li>Número de filas totales</li>
    <li>Número de columnas totales</li>
    <li>Índice de la imagen dentro del número total de imágenes</li>
</ul>

Es decir, si queremos mostrar una fila con dos imágenes, utilizaremos dos veces la función <code>subplot</code>, indicando el valor 1 para el número de filas y el valor 2 para el de columnas en ambas llamadas a la función. El índice de la imagen será 1 para la primera imagen y 2 para la segunda. Podemos verlo en el siguiente ejemplo: 


```python
image2 = imread('img/image2.jpg')

plt.figure(figsize=(10,8))
plt.subplot(1,2,1) # 1 - numrows, 2 - numcols, 1 - image index
plt.imshow(image1) 
plt.axis('off')
plt.subplot(1,2,2) # # 1 - numrows, 2 - numcols, 2 - image index
plt.imshow(image2)
plt.axis('off')
plt.show()
```

Y, si queremos mostrar un grid con cuatro imágenes, tendremos que indicar que el número de filas es 2 y el de columnas también es 2, como en el siguiente ejemplo:

```python
plt.figure()
plt.subplot(2,2,1) 
plt.imshow(image1) 
plt.axis('off') 
plt.subplot(2,2,2) 
plt.imshow(image2)
plt.axis('off')
plt.subplot(2,2,3) 
plt.imshow(image1)
plt.axis('off')
plt.subplot(2,2,4) 
plt.imshow(image2)
plt.axis('off')
plt.show()
```
